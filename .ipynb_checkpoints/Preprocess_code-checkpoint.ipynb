```python
def mask_df(df):
    mask1 = (df['Last BP(D)'].notnull() & df['Last BP(S)'].notnull() & df['Last BMI'].notnull())
    mask2 = (df['Encounter Diagnoses'].notnull() | df['Problem List'].notnull())
    return df[mask1 & mask2].copy()
```

```python
medex_input = "../Data/MedEx_input/"
def add_medex_prep(df, medex_input, dis_type):
    if not os.path.exists(medex_input + dis_type +"/"):
        os.makedirs(medex_input + dis_type +"/")
    for pid in df.index:
        f = open(medex_input + dis_type +"/" + "medex_" + str(pid) + ".txt", "w")
        input_ = df['Current Medications'][pid]
        try:
            f.write(input_)
        except:
            pass
        f.close()
```

```python
import re 
  
def remove(string): 
    pattern = re.compile(r'\s+') 
    return re.sub(pattern, '', string) 
```

```python
def medex_output_df(df, medex_output, dis_type):
    big_lst =[]
    for pid in df.index:
        path_file = medex_output + dis_type +"/"+ "medex_" + str(pid) + ".txt"
        if os.path.isfile(path_file):
            parsed = build_medex_obj(sys_url(path_file))
            lst_med = ''
            for med in parsed:
                lst_med += ' '+ remove(med[-1])
                big_lst.append(med[-1])
            df['Current Medications'][pid] = lst_med
    return df.copy(), big_lst
```

```python
df_healthy_new, list_ = medex_output_df(df_healthy, medex_output, 'healthy')
df_AD_new, list_ = medex_output_df(df_AD, medex_output, 'AD')
df_VD_new, list_ = medex_output_df(df_VD, medex_output, 'VD')
df_MCI_new, list_ = medex_output_df(df_MCI, medex_output, 'MCI')
df_PD_new, list_ = medex_output_df(df_PD, medex_output, 'PD')
```

#### Find the first 8 most frequent drugs for each sub-group

```python
def med_freq(med_subclass,vect):
    med_subclass[med_subclass > 0] = 1
    freq_med_s = med_subclass.sum(axis=0)
    top_freq = freq_med_s [np.argsort(freq_med_s )[::-1]]
    feature_med = np.array(vect.get_feature_names())
    top_med = feature_med[np.argsort(freq_med_s)[::-1]]
    return top_freq[:8],top_med[:8]
```

```python
med_freq_HA, med_name_HA = med_freq(m_med[df_all['target'] == 1],vectorizer)
med_freq_AD, med_name_AD = med_freq(m_med[df_all['target'] == 2],vectorizer)
med_freq_VD, med_name_VD = med_freq(m_med[df_all['target'] == 3],vectorizer)
med_freq_PD, med_name_PD = med_freq(m_med[df_all['target'] == 5],vectorizer)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    import itertools
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
confusion1, confusion2, df_score_rf_test = DementiaClassifier(X_train,y_train,rf).score_model(X_test,y_test)
```
